```python
class GAN_dataset(Dataset):
    
    def __init__(self, data_dir, latent_dim, n_lon, n_lat, m_lon, m_lat, date_range, variables, var_stds, mean_zarr_name, model_zarr_name, lon_buffers = [None, None], lat_buffers = [None, None], multiprocessing = True):
        self.data_dir = data_dir
        self.latent_dim = latent_dim
        self.n_lon = n_lon
        self.n_lat = n_lat
        self.m_lon = m_lon
        self.m_lat = m_lat
        self.lon_buffers = lon_buffers
        self.lat_buffers = lat_buffers
        self.date_range = date_range
        self.variables = variables
        self.var_stds = var_stds
        self.n_channels = len(variables)
        self.mean_zarr_name = mean_zarr_name
        self.model_zarr_name = model_zarr_name
        
        if not multiprocessing:
            self.ds_model = xr.open_dataset(self.data_dir + self.model_zarr_name, engine='zarr')
            self.ds_mean = xr.open_dataset(self.data_dir + self.mean_zarr_name, engine='zarr')
            
            i_lon_min = self.lon_buffers[0]
            i_lon_max = -self.lon_buffers[1]
            i_lat_min = self.lat_buffers[0]
            i_lat_max = -self.lat_buffers[1]
                
            self.ds_model = self.ds_model.isel(longitude = slice(i_lon_min, i_lon_max), latitude = slice(i_lat_min, i_lat_max), depth = 0, drop = True).sel(time=slice(self.date_range[0],self.date_range[1]), drop = True)
            self.ds_mean = self.ds_mean.isel(longitude = slice(i_lon_min, i_lon_max), latitude = slice(i_lat_min, i_lat_max), depth = 0, drop = True).sel(time=slice(self.date_range[0],self.date_range[1]), drop = True)
            
            self.N_lon = self.ds_model.dims['longitude']
            self.N_lat = self.ds_model.dims['latitude']
            self.N_time = self.ds_model.dims['time']
            
            # Generate isel arguments
            self.indexer = []
            self.mean_indexer = []
            np.random.seed(1)
            for t in range(self.N_time):
                lat_indices = np.random.choice(self.N_lat - self.n_lat + 1, self.m_lat, replace=False)
                lon_indices = np.random.choice(self.N_lon - self.n_lon + 1, self.m_lon, replace=False)
                for lat_start in lat_indices:
                    for lon_start in lon_indices:
                        self.indexer.append({
                            'latitude': slice(lat_start, lat_start + self.n_lat),
                            'longitude': slice(lon_start, lon_start + self.n_lon),
                            'time': t
                        })
                        self.mean_indexer.append({
                            'latitude': slice(lat_start, lat_start + self.n_lat),
                            'longitude': slice(lon_start, lon_start + self.n_lon)
                        })
            
            
        
    def __len__(self):
        return int(self.m_lon*self.m_lat*(self.date_range[1]-self.date_range[0]).days)
    
    def worker_init_fn(self, worker_id):
        self.ds_model = xr.open_dataset(self.data_dir + self.model_zarr_name, engine='zarr')
        self.ds_mean = xr.open_dataset(self.data_dir + self.mean_zarr_name, engine='zarr')

        i_lon_min = self.lon_buffers[0]
        if self.lon_buffers[1] is not None:
            i_lon_max = -self.lon_buffers[1]
        else:
            i_lon_max = None
        i_lat_min = self.lat_buffers[0]
        if self.lat_buffers[1] is not None:
            i_lat_max = -self.lat_buffers[1]
        else:
            i_lat_max = None

        self.ds_model = self.ds_model.isel(longitude = slice(i_lon_min, i_lon_max), latitude = slice(i_lat_min, i_lat_max), depth = 0, drop = True).sel(time=slice(self.date_range[0],self.date_range[1]), drop = True)
        self.ds_mean = self.ds_mean.isel(longitude = slice(i_lon_min, i_lon_max), latitude = slice(i_lat_min, i_lat_max), depth = 0, drop = True)

        self.N_lon = self.ds_model.dims['longitude']
        self.N_lat = self.ds_model.dims['latitude']
        self.N_time = self.ds_model.dims['time']

        # Generate isel arguments
        self.indexer = []
        self.mean_indexer = []
        np.random.seed(1)
        for t in range(self.N_time):
            lat_indices = np.random.choice(self.N_lat - self.n_lat + 1, self.m_lat, replace=False)
            lon_indices = np.random.choice(self.N_lon - self.n_lon + 1, self.m_lon, replace=False)
            for lat_start in lat_indices:
                for lon_start in lon_indices:
                    self.indexer.append({
                        'latitude': slice(lat_start, lat_start + self.n_lat),
                        'longitude': slice(lon_start, lon_start + self.n_lon),
                        'time': t
                    })
                    self.mean_indexer.append({
                        'latitude': slice(lat_start, lat_start + self.n_lat),
                        'longitude': slice(lon_start, lon_start + self.n_lon)
                    })
        

    def __getitem__(self, idx):
        
        invar = torch.from_numpy(np.random.randn(self.latent_dim))
        
        data_model = self.ds_model.isel(self.indexer[idx])
        data_mean = self.ds_mean.isel(self.mean_indexer[idx])
        
        outvar = np.zeros((self.n_channels, self.n_lat, self.n_lon))
        for v, var in enumerate(self.variables):
            outvar[v,] = (data_model[var] - data_mean[var])/self.var_stds[var]
        
        outvar = torch.from_numpy(outvar.astype(np.float32))
        outvar = torch.nan_to_num(outvar, nan=0.0)
        
        return invar, outvar
```

```python
dataset = GAN_dataset(data_dir = './input_data/', latent_dim = 64, n_lon = 128, n_lat = 128, m_lon = 4, m_lat = 4, date_range = [date(2019,1,1),date(2019,12,31)], variables = ['zos', 'thetao', 'so', 'uo', 'vo'], var_stds = var_stds, mean_zarr_name = 'glorys_gulfstream_means_zarr', model_zarr_name = 'glorys_gulfstream_zarr', lon_buffers = [3, None], lat_buffers = [None, 2], multiprocessing = True)
```

```python
batch_size = 25
n_cpus = 2
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True, num_workers=n_cpus, worker_init_fn = dataset.worker_init_fn,persistent_workers=True)
```

```python
for step, (x, y) in enumerate(dataloader):
    print(step)
```
